2. Write a Python program to create a plot to get a general Statistics of

3. Write a python program to create a graph to find relationship between 
the Culmen Length and Culmen Depth.

```python
sns.set_style('whitegrid')
sns.FacetGrid(df,hue='Species',height=4)\
.map(plt.scatter,'Culmen Length (mm)','Culmen Depth (mm)')\
.add_legend()
plt.show()
```

5. Write a Python program to find the correlation between variables of 
penguin data. Also create a heatmap using Seaborn to present their 
relations.

```python
import seaborn as sns
sns.FacetGrid(df,hue='Species',height=4)\
    .map(sns.distplot,'Culmen Length (mm)')\
    .add_legend()
plt.show()
```
